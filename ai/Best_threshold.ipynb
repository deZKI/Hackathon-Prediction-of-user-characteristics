```python
import logging
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from catboost import CatBoostClassifier, Pool


# Настройка логирования
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Загрузка данных
train_events = pd.read_csv('train_events.csv')
train_targets = pd.read_csv('train_targets.csv')

# Объединение данных по 'viewer_uid'
train_data = pd.merge(train_events, train_targets, on='viewer_uid', how='left')

# Извлечение дополнительных признаков из временных меток
for df in [train_data]:
    df['event_timestamp'] = pd.to_datetime(df['event_timestamp'])
    df['Hour'] = df['event_timestamp'].dt.hour
    df['Day'] = df['event_timestamp'].dt.day
    df['Month'] = df['event_timestamp'].dt.month
    df['Year'] = df['event_timestamp'].dt.year
    df['DayOfWeek'] = df['event_timestamp'].dt.dayofweek

# Определение категориальных признаков
categorical_features = [
    'region', 'ua_device_type', 'ua_client_type', 'ua_os',
    'ua_client_name', 'rutube_video_id'
]

# Кодирование категориальных признаков
categorical_mapping = {}
num_unique = {}
for col in categorical_features:
    unique_values = train_data[col].astype(str).unique().tolist()
    mapping = {val: idx + 1 for idx, val in enumerate(unique_values)}  # Начинаем с 1, чтобы 0 зарезервировать для 'no video'
    categorical_mapping[col] = mapping
    num_unique[col] = len(mapping) + 1  # +1 для неизвестных категорий
    train_data[col] = train_data[col].astype(str).map(mapping).fillna(0).astype(int)  # 0 для неизвестных

# Преобразование таргетов
train_data['sex'] = train_data['sex'].map({'female': 0, 'male': 1})

# Масштабирование числовых признаков
numerical_features = ['total_watchtime', 'Hour', 'Day', 'Month', 'Year', 'DayOfWeek']
scaler = StandardScaler()
train_data[numerical_features] = scaler.fit_transform(train_data[numerical_features])

# Группировка по 'viewer_uid' и подсчёт количества просмотренных видео
df_grouped = train_data.groupby('viewer_uid')['rutube_video_id'].apply(list).reset_index(name='videos')
df_grouped['len'] = df_grouped['videos'].apply(lambda x: len(x))

# Объединение с таргетами
merg = pd.merge(train_targets[['viewer_uid', 'age_class', 'sex']], df_grouped, on='viewer_uid', how='left')

# Обработка пользователей без просмотров
merg['len'] = merg['len'].fillna(0).astype(int)
merg['videos'] = merg['videos'].apply(lambda x: x if isinstance(x, list) else [])

# Преобразование таргетов
merg['sex'] = merg['sex'].map({'female': 0, 'male': 1})

# Инициализация LabelEncoder для age_class
label_encoder_age = LabelEncoder()
merg['age_class_encoded'] = label_encoder_age.fit_transform(merg['age_class'])

# Убедимся, что все необходимые признаки присутствуют
required_features = ['len'] + numerical_features
for feat in required_features:
    if feat not in merg.columns:
        merg[feat] = 0  # Заполнение отсутствующих признаков нулями или другим подходящим значением

# Вывод распределения количества просмотров
sns.histplot(merg['len'].to_list(), bins=30, kde=True)
plt.title('Распределение количества просмотренных видео')
plt.xlabel('Количество видео')
plt.ylabel('Количество пользователей')
plt.xlim([0, merg['len'].max()])
plt.show()

# Определение порогов
thresholds = [0, 5, 10, 15, 20, 25]

# Инициализация списков для сбора метрик
metrics_sex = {'threshold': [], 'accuracy': [], 'f1_score': [], 'report': []}
metrics_age = {'threshold': [], 'accuracy': [], 'f1_score': [], 'report': []}

# Разделение данных на обучающую и тестовую выборки **до** применения порогов
# Это обеспечит, что тестовая выборка всегда содержит пользователей с len=0
X_full = merg[['len'] + numerical_features + ['videos']]
y_sex_full = merg['sex']
y_age_full = merg['age_class_encoded']

# Функция для подготовки признаков: разделение видео на отдельные колонки
def prepare_video_features(videos, max_len=64):
    padded = videos[:max_len] + [0]*(max_len - len(videos))  # 0 зарезервирован для 'no video'
    return padded

# Применение функции к видео
merg['videos_padded'] = merg['videos'].apply(lambda x: prepare_video_features(x, max_len=64))
video_cols = [f'video_{i+1}' for i in range(64)]
video_df = pd.DataFrame(merg['videos_padded'].tolist(), columns=video_cols)

# Объединение признаков
X_full = pd.concat([merg[['len'] + numerical_features], video_df], axis=1)

# Разделение на обучающую и тестовую выборки
X_train_full, X_test_full, y_train_sex_full, y_test_sex_full, y_train_age_full, y_test_age_full = train_test_split(
    X_full, y_sex_full, y_age_full, test_size=0.2, random_state=42, stratify=y_sex_full
)

# Итерация по порогам
for threshold in thresholds:
    logging.info(f"\n=== Порог: {threshold} ===")
    
    # Фильтрация обучающей выборки по порогу
    X_train_filtered = X_train_full[X_train_full['len'] >= threshold].reset_index(drop=True)
    y_train_sex_filtered = y_train_sex_full[X_train_full['len'] >= threshold].reset_index(drop=True)
    y_train_age_filtered = y_train_age_full[X_train_full['len'] >= threshold].reset_index(drop=True)
    
    # Проверка, что после фильтрации осталось достаточно данных
    if len(X_train_filtered) == 0:
        logging.warning(f"Нет данных после фильтрации с порогом {threshold}")
        continue
    
    # Определение категориальных признаков для CatBoost (все video_* колонки)
    cat_features = video_cols
    
    # Обучение модели для пола
    logging.info("Обучение модели CatBoost для предсказания пола (sex)...")
    model_cat_sex = CatBoostClassifier(
        iterations=1000,
        learning_rate=0.1,
        depth=6,
        loss_function='Logloss',
        eval_metric='AUC',
        random_seed=42,
        verbose=100,
        early_stopping_rounds=50
    )
    
    # Обучение модели
    model_cat_sex.fit(
        X_train_filtered, y_train_sex_filtered,
        eval_set=(X_test_full, y_test_sex_full),
        cat_features=cat_features,
        use_best_model=True
    )
    
    # Предсказания для пола
    y_pred_sex = model_cat_sex.predict(X_test_full)
    y_proba_sex = model_cat_sex.predict_proba(X_test_full)[:, 1]
    
    # Вычисление метрик для пола
    acc_sex = accuracy_score(y_test_sex_full, y_pred_sex)
    f1_sex = f1_score(y_test_sex_full, y_pred_sex, average='weighted')
    report_sex = classification_report(y_test_sex_full, y_pred_sex, target_names=['female', 'male'])
    
    metrics_sex['threshold'].append(threshold)
    metrics_sex['accuracy'].append(acc_sex)
    metrics_sex['f1_score'].append(f1_sex)
    metrics_sex['report'].append(report_sex)
    
    logging.info(f"Sex - Accuracy: {acc_sex:.4f}, F1-Score: {f1_sex:.4f}")
    print("\nClassification Report (Sex):")
    print(report_sex)
    
    # Визуализация матрицы ошибок для пола
    cm_sex = confusion_matrix(y_test_sex_full, y_pred_sex)
    disp_sex = ConfusionMatrixDisplay(confusion_matrix=cm_sex, display_labels=['female', 'male'])
    disp_sex.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix - Sex (Threshold: {threshold})')
    plt.show()
    
    # Обучение модели для возраста
    logging.info("Обучение модели CatBoost для предсказания возраста (age_class)...")
    model_cat_age = CatBoostClassifier(
        iterations=1000,
        learning_rate=0.1,
        depth=6,
        loss_function='MultiClass',
        eval_metric='MultiClass',
        random_seed=42,
        verbose=100,
        early_stopping_rounds=50
    )
    
    # Обучение модели
    model_cat_age.fit(
        X_train_filtered, y_train_age_filtered,
        eval_set=(X_test_full, y_test_age_full),
        cat_features=cat_features,
        use_best_model=True
    )
    
    # Предсказания для возраста
    y_pred_age = model_cat_age.predict(X_test_full)
    y_proba_age = model_cat_age.predict_proba(X_test_full)
    
    # Вычисление метрик для возраста
    acc_age = accuracy_score(y_test_age_full, y_pred_age)
    f1_age = f1_score(y_test_age_full, y_pred_age, average='weighted')
    report_age = classification_report(y_test_age_full, y_pred_age, target_names=label_encoder_age.classes_.astype(str))
    
    metrics_age['threshold'].append(threshold)
    metrics_age['accuracy'].append(acc_age)
    metrics_age['f1_score'].append(f1_age)
    metrics_age['report'].append(report_age)
    
    logging.info(f"Age - Accuracy: {acc_age:.4f}, F1-Score: {f1_age:.4f}")
    print("\nClassification Report (Age):")
    print(report_age)
    
    # Визуализация матрицы ошибок для возраста
    cm_age = confusion_matrix(y_test_age_full, y_pred_age)
    disp_age = ConfusionMatrixDisplay(confusion_matrix=cm_age, display_labels=label_encoder_age.classes_.astype(str))
    disp_age.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix - Age Class (Threshold: {threshold})')
    plt.show()

# Визуализация метрик по порогам
# Преобразование метрик в DataFrame для удобства визуализации
metrics_sex_df = pd.DataFrame(metrics_sex)
metrics_age_df = pd.DataFrame(metrics_age)

# Визуализация метрик для пола
plt.figure(figsize=(12, 6))
plt.plot(metrics_sex_df['threshold'], metrics_sex_df['accuracy'], marker='o', label='Accuracy')
plt.plot(metrics_sex_df['threshold'], metrics_sex_df['f1_score'], marker='s', label='F1-Score')
plt.title('Метрики для предсказания пола по порогу')
plt.xlabel('Порог минимального количества просмотров')
plt.ylabel('Метрика')
plt.legend()
plt.grid(True)
plt.show()

# Визуализация метрик для возраста
plt.figure(figsize=(12, 6))
plt.plot(metrics_age_df['threshold'], metrics_age_df['accuracy'], marker='o', label='Accuracy')
plt.plot(metrics_age_df['threshold'], metrics_age_df['f1_score'], marker='s', label='F1-Score')
plt.title('Метрики для предсказания возраста по порогу')
plt.xlabel('Порог минимального количества просмотров')
plt.ylabel('Метрика')
plt.legend()
plt.grid(True)
plt.show()
```
